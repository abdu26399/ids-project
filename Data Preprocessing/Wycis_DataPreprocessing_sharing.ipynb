##Checking Object datatypes which has categories

```python
#Find all the object datatypes in the dataframe
objcategories = df_KddDS.select_dtypes(include='O').keys()
objcategories
```

```python
#Save a copy of columns with object datatype into the new dataframe
new_df = pd.read_csv('/content/kddcup99_csv.csv',usecols=objcategories) 
new_df.head()
```

##One hot encoding

```python
# for all categorical variables we selected encode 1/0
def top_x(df2,variable,top_x_labels):
    for label in top_x_labels:
        df2[variable+'_'+label] = np.where(data[variable]==label,1,0)
# read the data for the required catageroues again
data = pd.read_csv('/content/kddcup99_csv.csv',usecols=objcategories) 
#encode to one-hot format for every object category
for i in objcategories:
    categories = [x for x in new_df[i].value_counts().sort_values(ascending=False).index]
    top_x(data,i,categories)
# display data
data.head()
```

```python
#dropping these columns to eliminate duplicates after concatination
data.drop(['protocol_type', 'service', 'flag', 'label'],axis=1,inplace=True)
```
